```python
from tensorflow.keras.models import load_model
```

```python
model=load_model(r'C:\Users\Lenovo\Downloads\generator_all.h5')
```

```python
model
```

```python
import numpy as np
import matplotlib.pyplot as plt
NOISE_DIM=100
noise = np.random.normal(0,1,size=(36, NOISE_DIM))
generated_img=model.predict(noise)
plt.figure(figsize=(10,10))
for i in range(36):
  plt.subplot(6,6,i+1)
  plt.imshow(generated_img[i].reshape(28,28), cmap='binary')
  plt.axis('off')
plt.show()
```
